```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
from Hyundai_module import crawl_three_page, crawl_one_page, move_to_next_page

chromedriver_autoinstaller.install()
driver = webdriver.Chrome()
action = ActionChains(driver)
```

```python
# 크롤링할 웹 페이지 URL
url = 'https://www.kia.com/kr/customer-service/center/faq#none'
driver.get(url)
time.sleep(3)
```

```python
# 차량 구매 관련 FAQ 구매
results = []
# 기다릴 최대 시간 설정
wait = WebDriverWait(driver, 6)
```

```python
try:
    for i in range(3,9):
        button = driver.find_element(By.XPATH, f'//*[@id="tab-list"]/li[{i}]/button/span')
        button.click()
        tap = button.text
        time.sleep(5)
        try:
            for i in range(1,5):
                if i == 2:
                    button_0 = driver.find_element(By.XPATH, f'//*[@id="contents"]/div/div[3]/div/div/div[4]/div/button')
                    button_0.click()
                    time.sleep(5)
                elif i > 2:
                    button_0 = driver.find_element(By.XPATH, f'//*[@id="contents"]/div/div[3]/div/div/div[4]/div/button[2]')
                    button_0.click()
                    time.sleep(5)
                for i in range(1,6):
                    button_0 = driver.find_element(By.XPATH, f'//*[@id="contents"]/div/div[3]/div/div/div[4]/div/ul/li[{i}]/a')
                    button_0.click()
                    time.sleep(3)
                    try:
                        for i in range(15):
                            button_1 = driver.find_element(By.XPATH, f'//*[@id="accordion-item{i}-button"]/span[1]')
                            button_1.click()
                    except:
                        print("페이지 내의 모든 Question탭 활성화 완료")
                    time.sleep(3)
                    try:
                        for i in range(15):
                            result = {}
                            button_1 = driver.find_element(By.XPATH, f'//*[@id="accordion-item{i}-button"]/span[1]')
                            result['question'] ='[' + tap + ']' + button_1.text
                            result2 = {}
                            button_2 = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/div[2]/div/div/div[3]/div/div/div[3]/div/div/div[{i+1}]/div/div/div/div')
                            result['answer'] = button_2.text
                            result['tap'] = tap
                            results.append(result)
                    except:
                        print("페이지 내의 모든 Question, Answer 크롤링 완료")
        except:
            print("카테고리탭의 모든 페이지 크롤링 완료")
except:
    print("모든 카테코리 크롤링 완료")
```

```python
import re

# 제어 문자를 제거하는 함수
def remove_control_characters(text):
    if isinstance(text, str):
        # 정규 표현식을 사용하여 제어 문자를 제거
        return re.sub(r'[\x00-\x1F\x7F]', '', text)
    return text

# CSV 파일 읽기
df = pd.read_csv('kia_faq3.csv')

# 데이터프레임의 모든 셀에 대해 제어 문자 제거
df = df.applymap(remove_control_characters)

# 수정된 데이터프레임을 새로운 CSV 파일로 저장
df.to_csv('kia_faq3_cleaned.csv', index=False)
```
